```python
import pandas as pd
import numpy as np
import os 
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
ruta=os.path.join(os.getcwd(),'ai_job_dataset.csv')
df=pd.read_csv(ruta)
df.head(10)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 999
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.title('Distribucion del salario',fontsize=15)
sns.histplot(df.salary_usd,kde=True, color= "Green", edgecolor= "black")
plt.xlabel('Salario en (USD)')
plt.ylabel('Frecuencia')
```

```python
plt.figure(figsize=(10,5))
plt.title('Experiencia vs Salario',fontsize=15)
sns.boxplot(data=df, x='experience_level', y='salary_usd',hue='company_size', order=['EN', 'MI', 'SE','EX'], hue_order=['S','M','L'])
plt.xlabel('Experiencia')
plt.ylabel('Salario en (USD)')
```

S (Small <50), M (Medium 50-250), L (Large >250)

```python
# Se realiza un agrupamiento de los datos para mostrar el salario promedio por cada industria
industry_salary=df.groupby(by='industry')['salary_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
sns.barplot(x=industry_salary.values, y=industry_salary.index, hue=industry_salary.index)
plt.xlabel('Salario promedio en (USD)')
plt.ylabel('Industria')
plt.title('Salario promedio por industria')
plt.xlim(industry_salary.values.min()-500, industry_salary.values.max()+500)
```

```python
(industry_salary['Consulting']-industry_salary['Gaming'])/12
```

```python
type_salary=df.groupby(by='employment_type')['salary_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=type_salary.index, y=type_salary.values, hue=type_salary.values, order=['PT','FL','CT','FT'], palette='dark:green')
plt.xlabel('Tipo de trabajo')
plt.ylabel('Salario promedio en (USD)')
plt.title('Salario promedio por tipo')
plt.ylim(type_salary.min()-500, type_salary.max()+500)
```

```python
df['required_skills']
```

```python
''' 
Para procesar la habilidades mas demandas entonces realizamos un preprocesamiento a la columna required_skills,
Se evidencia que esta columna trae valores repetidos, por cada celda, me gustaria visualizar por individual cuales son las habilidades mas repetidas, para ellos
eliminamos los espacios en blanco, dividimos el contenido de cada celda por comas, y agregamos explode para convertir cada elemento en una lista separada
'''
all_reqs = df["required_skills"].dropna().str.split(",").explode()
all_reqs = all_reqs.str.strip() # eliminamos espacios en blanco al principio o al final 
req_counts = all_reqs.value_counts() #Conteo por elemento

wordcloud = WordCloud(width=1000, height=400, background_color='white').generate_from_frequencies(req_counts)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Habilidades ms demandadas", fontsize=16)
plt.show()
```

```python
company_salary=df.groupby('company_location').agg(salary_mean=('salary_usd', 'mean'), count_jobs=('job_id','count')).sort_values(by='salary_mean',ascending=False).reset_index()

fig, ax1 = plt.subplots(figsize=(14, 6))

# Barras: salario promedio
sns.barplot(data=company_salary, x='company_location', y='salary_mean',hue='company_location', ax=ax1, palette='Greens_r')
ax1.set_ylabel('Salario promedio (USD)')
ax1.tick_params(axis='y')
ax1.set_xlabel('Pas')
ax1.set_title('Salario promedio y nmero de empleos por pas')
plt.xticks(rotation=45)

# Lnea: cantidad de empleos (segundo eje Y)
ax2 = ax1.twinx()
sns.lineplot(data=company_salary,x='company_location',y='count_jobs',color='darkorange',marker='o',linewidth=2,ax=ax2)
ax2.set_ylabel('Cantidad de empleos', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

plt.tight_layout()
plt.show()
```

```python
df.groupby(['job_title','remote_ratio'])['job_title'].count()
# Calcular el total de empleos por cada job_title (sin remote_ratio)
df_grouped = df.groupby(['job_title','remote_ratio'])['job_title'].count().reset_index(name='num_empleos')

total_empleos = df['job_title'].value_counts().rename_axis('job_title').reset_index(name='total_empleos')

# Unir al df_grouped
df_grouped = df_grouped.merge(total_empleos, on='job_title')
df_grouped
```

```python
df.groupby(by=['remote_ratio']).agg(salary_mean=('salary_usd','mean'),experience_mean=('years_experience','mean'))
```

```python
# Agrupar salario medio por cargo y modalidad remota
df_grouped = df.groupby(['job_title', 'remote_ratio']).agg(salary_mean=('salary_usd', 'mean')).reset_index()

# Calcular el rango (mx - mn) por cargo
salary_diff = df_grouped.groupby('job_title')['salary_mean'].agg(lambda x: x.max() - x.min()).sort_values(ascending=False)

# Seleccionar los N cargos con mayor diferencia (ej: top 6)
top_jobs = salary_diff.head(6).index

filtered = df[df['job_title'].isin(top_jobs)]
plt.figure(figsize=(12,6))
sns.boxplot(
    data=filtered,
    x='job_title',
    y='salary_usd',
    hue='remote_ratio',
    palette='Set2'
)
plt.title('Impacto del trabajo remoto en los salarios (Top cargos con mayor variacin)')
plt.xlabel('Cargo')
plt.ylabel('Salario en USD')
plt.xticks(rotation=45)
plt.legend(title='Remote %')
plt.tight_layout()
plt.show()
```

```python
corr_matrix =df[['salary_usd','remote_ratio', 'years_experience', 'job_description_length','benefits_score']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='YlGnBu', linewidths=0.5)
plt.title("Matriz de correlacin")
plt.tight_layout()
plt.show()
```
